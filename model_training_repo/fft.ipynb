```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1250
dt = 5.0 / 1250

data = []

with open('./data/tinyml_contest_data_training/S91-AFt-191.txt') as file:
    for line in file:
        data.append(float(line.strip()))

data = torch.tensor(data)
```

```python
plt.rcParams["figure.figsize"] = (30,4)
plt.xticks( range(0,130,10) )

freq = (1/(dt*n)) * np.arange(n)
fhat = torch.fft.fft(data)
psd = abs(fhat) / n
L = np.arange(0, np.floor(n/2), dtype='int')
plt.xlim(freq[L[0]], freq[L[-1]])
plt.plot(freq, psd)
plt.show()

plt.xticks([0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5])
time = [5*(x/1250) for x in range(len(data))]
plt.xlim(0, 5)
plt.plot(time, data)
plt.show()
```
